```python
player = str(input('your name: ')) # Asks player to enter the name.
print('Hi! ',player, "Get ready for Number Guessing Game.") # wellcoming the player with his/her name in the game.
import random # Generating or manuplating the random numerics with the random module.
device_estimate = random.randint(1,100) # returns the random numerics from given range 1 to 100.
calculate = 0 # Initializing the counting of iteration of a codes. 
#print(device_estimate) # printing the random ineteger which is generated by the device # just for reference. 

while True: #initiating loop, beacuse the player have to estimate an integer.
    # ask the player to enter an integer, which is between 1 to 100 or press 0 to shutdown the game in the middle of the game. 
    player_estimate = int(input("Estimate the numeric between '1' and '100' or enter '0' to shutdown =" ))
    calculate += 1 # will calculate the numeric of estimation.
    print('attempt',calculate, '=', player_estimate)
    
    if player_estimate !=0: # assuming that the player will enter the digit apart from zero.
        if device_estimate != player_estimate: # assuming that the entered numeriv by the player is not matching the estimated numeric by the device.
            if device_estimate < player_estimate: # comapring the numeric entered by the player and estimated by a computer.
                print('estimated value is above limit')
            else:
                print('estimated value is under limit')
            continue # depecting to continue the loop.
        else: # assuming that the entered numeric by the player matches with the estimated numeric by the device.
             print('oo! Yeah! You have estimated.') # telling the player that they have estimated the exact numeric.
             your_achievement = int(1000/calculate) # calculting the overall achievements of the player.
             print(player,'achievements is =',your_achievement) # returns to the achievement value.
             break # to stop the loop, because the player has estimated correct estimted numeric. 
    else: # assuming that the player has entered the value zero as he/she wants to shutdown the game. 
        Shutdown = str(input("Really wanted to stand up ? 'Yes' or 'No' = ")) # asking the player, whether he/she wanted to shutdown the game? 
        if Shutdown == 'No': # assuming that the player wants to continue the game further.
          continue # continuation of the loop. 
        elif Shutdown == 'Yes': # assuming that the player wants to stop the game.
          print('So Sorry! better luck next time') # giving feedback to the player, as he/she has failed. 
          print('Thank You!') # saying goodbye to the player. 
          break # stopping the loop here, because the player wants to shutdown the game. 
          
    


```
